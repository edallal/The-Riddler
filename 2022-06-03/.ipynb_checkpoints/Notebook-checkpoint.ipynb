```python
import matplotlib.pyplot as plt
import matplotlib.patches as pat
import pandas as pd
import numpy as np
```

```python
# Create explanation plot
import math
plt.axes()
desert = plt.Line2D((-2, 2), (0, 0), lw=4, c='y')
plt.gca().add_line(desert)
line1 = plt.Line2D((1, 3), (0, 2), ls = '--')
plt.gca().add_line(line1)
arc1 = pat.Arc((1,0), 1, 1, 0, 0, 45, label = u"\u03b1")
plt.gca().add_patch(arc1)
line2 = plt.Line2D((0, 1), (0, 2), ls = '--')
plt.gca().add_line(line2)
arc2 = pat.Arc((0,0), 1, 1, 0, 0, 180*math.atan(2)/math.pi, label = u"\u03b2")
plt.gca().add_patch(arc2)
line3 = plt.Line2D((-1, -1.5), (0, 2), ls = '--')
plt.gca().add_line(line3)
arc3 = pat.Arc((-1,0), 1, 1, 0, 0, 180*(math.pi-math.atan(4))/math.pi, label = u"\u03b3")
plt.gca().add_patch(arc3)
plt.axis('scaled')
plt.savefig('Angles.png')
```

```python
Nmax = 20
N = list(range(1, Nmax))
Y = list(range(1, Nmax))
for n in range(2, Nmax):
    Y[n-1] = Y[n-2]*(2*n-1)/(n*n)
plt.plot(N, Y)
plt.yscale('log')
plt.xlabel('N')
plt.ylabel('Prob')
plt.title('Probability of no Intersections')
plt.savefig('Prob.png')
```
